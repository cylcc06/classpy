## 使用汽车品牌评价矩阵
        数据为10个汽车品牌之间相似性的用户评价

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA, FactorAnalysis, FastICA
from sklearn import preprocessing
import matplotlib.pyplot as plt
import os
%matplotlib inline

# os.chdir('Q:/data')
pd.set_option('display.max_columns', None)
```

```python
from sklearn.manifold import MDS
```

```python
car_df = pd.read_csv('CAR_BRANDS.csv', skipinitialspace=True)
car_df
```

整理表

```python
car_filled = car_df.fillna(0)
car_list = [car_filled.iloc[10*i: 10*i+10] for i in range(5)]
car_list
```

```python
from functools import reduce

car_list1 = map(lambda x: x.values + x.values.T, car_list)
car_sum = reduce(lambda x,y: x + y, car_list1)
car_sum
```

计算不同的q值对应的应力值

```python
def cal_stress(array, q):
    mds = MDS(n_components=q, dissimilarity='precomputed', random_state=123)
    mds.fit_transform(array)
    return (q, mds.stress_)

stress_q = []
for i in range(1,10):
    stress_q.append(cal_stress(car_sum, i))
stress_q
```

```python
plt.plot(np.array(stress_q).T[0], np.array(stress_q).T[1])
plt.show()
```

选择q=2，计算MDS

```python
car_mds = MDS(n_components=2, dissimilarity='precomputed', random_state=1)
car_mds.fit_transform(car_sum)
car_mds.embedding_
```

使用MDS计算新距离矩阵

```python
from sklearn.neighbors import DistanceMetric

dm = DistanceMetric.get_metric('euclidean')
dist = dm.pairwise(car_mds.embedding_)
pd.DataFrame(dist)
```

新距离矩阵与原矩阵的差

```python
diff = car_sum - dist
pd.DataFrame(diff)
```

计算新距离矩阵相对于原始矩阵的损失百分比

```python
ss_dist = lambda matrix: np.sqrt(np.power(matrix, 2).sum())

base = ss_dist(dist)
divider = ss_dist(diff)
loss = divider / base
print('Loss = %5.2f'  %loss)
```

绘制感知图

```python
plt.figure(figsize=(6, 6))
plt.ylim(-30, 30)
plt.xlim(-30, 30)
x = car_mds.embedding_[:, 0]
y = car_mds.embedding_[:, 1]
plt.scatter(x, y)
for a, b, s in zip(x, y, car_df.columns):
    plt.text(a, b, s, fontsize=14)
plt.show()
```
